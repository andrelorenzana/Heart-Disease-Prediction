```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('heart.csv')
train = df.sample(frac=0.8).reset_index(drop=True)
test = pd.concat([df,train]).drop_duplicates(keep=False).reset_index(drop=True)
train.to_csv('train.csv', index=False)
test.to_csv('test.csv', index=False)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(), ax=ax, annot=True)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.regplot(x='chol', y='target', ax=ax, data=df)
ax.set_title('Cholesterol vs Heart Disease')
plt.show()

```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.regplot(x='thalach', y='target', ax=ax, data=df)
ax.set_title('Thalach vs Heart Disease')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.bar(x='age', y='target', ax=ax, data=df)
ax.set_title('Age vs Heart Disease')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.regplot(x='oldpeak', y='target', ax=ax, data=df)
ax.set_title('Old Peak vs Heart Disease')
plt.show()
```

```python
train_df, test_df = train_test_split(df, test_size=0.1)
```

```python
x_train = train_df[train_df.columns.drop('target')]
y_train = train_df['target']
x_test = test_df[test_df.columns.drop('target')]
y_test = test_df['target']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
model.score(x_test, y_test)
```
